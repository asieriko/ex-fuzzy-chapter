# Fuzzy Sets

```python
from ex_fuzzy.fuzzy_sets import triangularFS, FS, gaussianFS

fsga = gaussianFS("Gaussian FS", [1.5,1], [0, 5])
fstri = triangularFS("Triangular FS", [1, 2,2, 3], [0, 5])
fstra = FS("Trapezoidal FS", [2, 3, 4, 5], [0, 5])
```

```python
from ex_fuzzy.fuzzy_sets import FS
from ex_fuzzy.fuzzy_sets import fuzzyVariable
from ex_fuzzy.vis_rules import plot_fuzzy_variable

fuzzy_variable = fuzzyVariable("Fuzzy Variable", [fsga, fstri, fstra])
plot_fuzzy_variable(fuzzy_variable)
```

```python
from ex_fuzzy.fuzzy_sets import IVFS, gaussianIVFS
ivfs = IVFS("high", [1.0,2.0,3.0,4.0],[0.5,1.5,3.5,4.5],  [0,5], 0.9)

from ex_fuzzy.vis_rules import plot_fuzzy_variable

ivfs_variable = fuzzyVariable("Interval Valued Fuzzy Set", [ivfs])
plot_fuzzy_variable(ivfs_variable)
```

# Fuzzy Decision Trees

```python
from ex_fuzzy.tree_learning import FuzzyCART
from ex_fuzzy.fuzzy_sets import FUZZY_SETS
from ex_fuzzy.utils import construct_partitions
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X, y = iris.data, iris.target


precomputed_partitions = utils.construct_partitions(
    X, FUZZY_SETS.t1, n_partitions=3
)

tree = FuzzyCART(precomputed_partitions)
tree.fit(X,y)
```

```python
tree.print_tree()
```

```python
y_predicted = tree.predict(X[0])
print(f"The predicted class is {y_predicted.item()}")
```
